```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("data-export (1).csv")
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
df.head()
```

```python
df = df.drop(index=0).reset_index(drop = True )
df. columns = ['channel group', 'DateHour','user','Sessions', 'Engaged Sesssions', 'Avg engagement time','Engaged Session per user','Event per session', 'Engagement rate', 'Event count']
```

```python
df.head()
```

```python
df.info()
```

```python
df['DateHour'] = pd.to_datetime(df['DateHour'], format="%Y%m%d%H", errors="coerce")
df.info()
```

```python
df.head()
```

```python
numaric_col = df.columns.drop(['channel group','DateHour'])
df[numaric_col]= df[numaric_col].apply(pd.to_numeric, errors = 'coerce')
df.info()
```

```python
df['Hour']= df['DateHour'].dt.hour 
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Session and user over time 

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Sessions","user"]].sum().plot(ax=plt.gca())
plt.title("Session and user over time")
plt.xlabel("DateHour") 
plt.ylabel("Count")
plt.show()
```

## Total users by channel

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x="channel group", y="user", estimator=np.sum, palette="viridis" )
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

## AVG Engagement time by channel

```python
plt.figure(figsize=(8,5))
sns.barplot(data= df, x="channel group", y="Avg engagement time", estimator=np.mean, palette="magma")
plt.title("AVG Engagement time by channel")
plt.xticks(rotation=45)
plt.show()
```

## Engagement rate distrubuted by channel"

```python
plt.figure(figsize=(8,5))
sns.boxplot(data= df, x="channel group", y="Engagement rate", palette="coolwarm")
plt.title("Engagement rate distrubuted by channel")
plt.xticks(rotation=45)
plt.show()
```

## engaged vs non engaged session

```python
df.head()
```

```python
session_df= df.groupby ("channel group")[["Sessions","Engaged Sesssions"]].sum().reset_index()
session_df["non-engaged"] = session_df["Sessions"]-session_df["Engaged Sesssions"]
session_df_melted= session_df.melt(id_vars="channel group", value_vars=["Engaged Sesssions","non-engaged"])

plt.figure(figsize=(8,5))
sns.barplot(data= session_df_melted, x="channel group",y= "value", hue= "variable")
plt.title("engaged vs non engaged session")
plt.xticks(rotation=45)
plt.show()
```

## traffic by hour and channel

```python
heatmap_data= df.groupby (["Hour", "channel group"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data,cmap="YlGnBu", linewidth=5,annot=True,fmt='.0f')
plt.title("traffic by hour and channel")
plt.xlable("channel group")
plt.ylable("Hour of Day")
plt.show()
```

## engagement rate vs sessions over time

```python
df.head()
```

```python
df_plot= df.groupby ("DateHour")[["Engagement rate", "Sessions"]].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(df_plot["DateHour"],df_plot["Engagement rate"],label= "Engagement rate",color="green")
plt.plot(df_plot["DateHour"],df_plot["Sessions"],label= "Sessions",color="blue")
plt.title("engagement rate vs sessions over time")
plt.xlable("DateHour")
plt.legend()
plt.grid(True)
plt.show()
```
